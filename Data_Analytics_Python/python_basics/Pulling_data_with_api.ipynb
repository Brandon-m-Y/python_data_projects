```python
print("We are only converting from the most current conversion rates")
print("We are currently only converting from the EURO to other currencies")
curr = input("Convert to (currency): ")
quant = float(input("How many Euros are you converting to {}: ".format(curr)))

# Add "+ ?base +" before "&symbols" to change base currency when allowed by API

url = "http://api.exchangeratesapi.io/v1/latest?access_key=4d55f21784c7c474f567ed16700b7766" + "&symbols=" + curr

response = requests.get(url)

if(response.ok is False):
    print("\nError {}:".format(response.status_code))
    print(response.json()["error"])

else:
    data = response.json()
    rate = data['rates'][curr]

    result = quant * rate

    print("\n{0} {1} is equal to {2} {3}, based upon exchange rates for right now. ".format(quant, "EURO", result, curr ))
```
