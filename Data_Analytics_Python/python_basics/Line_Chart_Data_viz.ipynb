```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_spx_ftse_00_10 = pd.read_csv("C:\\Users\\brand\\Downloads\\Data Analyst Course\\S34 - Data Visualization\\L16\\line_chart_data.csv")
```

```python
df_spx_ftse_00_10.head(10)
```

```python
df_spx_ftse_00_10['new date'] = pd.to_datetime(df_spx_ftse_00_10['Date'])
```

```python
df_spx_ftse_00_10['new date']
```

```python
labels = ['S&P 500', 'FTSE 100']
plt.figure(figsize = (20, 8))
plt.plot(df_spx_ftse_00_10['new date'], df_spx_ftse_00_10['GSPC500'])
plt.plot(df_spx_ftse_00_10['new date'], df_spx_ftse_00_10['FTSE100'])
plt.title('S&P vs FTSE Returns (2000 - 2010)', fontsize = 14, fontweight = 'bold')
plt.ylabel("Returns", fontsize = 14, fontweight = 'bold')
plt.xlabel("Date", fontsize = 14, fontweight = 'bold')
plt.legend(labels = labels, fontsize = 'large')
plt.show()
```

```python
df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10['new date'] >= '2000-07-01') &
                                      (df_spx_ftse_00_10['new date'] <= '2000-12-31')]
```

```python
df_spx_ftse_H2_08
```
