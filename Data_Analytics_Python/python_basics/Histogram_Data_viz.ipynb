```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("C:\\Users\\brand\\Downloads\\Data Analyst Course\\S34 - Data Visualization\\L20\\histogram_data.csv")
df_real_estate.head(10)
```

```python
sns.set_style('white')
plt.figure(figsize = (8, 6))
plt.hist(df_real_estate['Price'], 
         bins = 8,
         color = '#108A99')

plt.title('Distribution of Real Estate Prices', fontsize = 14, fontweight = 'bold')
plt.xlabel("Price in (000' $)", fontsize = 14, fontweight = 'bold')
plt.ylabel("Num Properties", fontsize = 14, fontweight = 'bold')
sns.despine()
plt.show()
```
