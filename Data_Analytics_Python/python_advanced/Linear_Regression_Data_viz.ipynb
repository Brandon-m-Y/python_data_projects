```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_expenditure = pd.read_csv("C:\\Users\\brand\\Downloads\\Data Analyst Course\\S34 - Data Visualization\\L29\\scatter_plot_ii.csv")
df_ad_expenditure.head(10)
```

```python
#plt.figure(figsize = (10, 8))
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(data = df_ad_expenditure,
            x = 'Budget',
            y = 'Sales',
            scatter_kws = {'color': 'k'},
            line_kws = {'color': 'red'})


plt.xlabel("Ad Expenditure in (000's $)", fontsize = 14, fontweight = 'bold')
plt.ylabel("Sales in (000's Units)", fontsize = 14, fontweight = 'bold')
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, fontweight = 'bold')
plt.show()
```

```python
sns.lmplot(x = 'Budget',
           y = 'Sales',
           data = df_ad_expenditure,
           height = 10,
           scatter_kws = {'color': 'k'},
           line_kws = {'color': 'red'})


plt.xlabel("Ad Expenditure in (000's $)", fontsize = 14, fontweight = 'bold')
plt.ylabel("Sales in (000's Units)", fontsize = 14, fontweight = 'bold')
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, fontweight = 'bold')
plt.show()
```
